```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def convertColumnToInt(df):
  for x in df.keys():
    df[x]=df[x].astype(int)
```

```python
heart_disease_df = pd.read_csv("/content/HEART_DISEASE_DF.csv")
heart_disease_df
```

```python
no_missing_values = heart_disease_df[(heart_disease_df['max_hr']!='-') & (heart_disease_df['restingbp']!='-') &(heart_disease_df['chol']!='-')&(heart_disease_df['blood_sugar']!='-')]
no_missing_values.describe()
```

```python
num = 0
count = 0
for x in no_missing_values.blood_sugar :
  if x =="1.0" or x =='1':
    num+=1
  else:
    num+=0
  count+=1

print(f'The average blood sugar is: {num/count}')
```

```python
sugar_age_plot = sns.violinplot(data=heart_disease_df[heart_disease_df['blood_sugar']!='-'],x='age',hue='blood_sugar',palette='magma').set_title('Age Distribution Separated by Blood Sugar Levels');
```

```python
diagnosis_age_plot = sns.violinplot(data=no_missing_values,x='age',hue='diagnosis',palette='crest').set_title('Age Distribution Separated by Heart Disease Diagnosis');
```

```python
palette_color = sns.color_palette('crest')
outward_piece = [0.05, 0, 0, 0, 0]
plt.title('Percentage of Heart Disease in the Population')
plt.pie(no_missing_values.diagnosis.value_counts(),labels=no_missing_values.diagnosis.unique(), colors=palette_color, autopct='%.0f%%',explode=outward_piece)
plt.show();
```

```python
no_heart_disease_data = no_missing_values[no_missing_values['diagnosis']==0]
```

```python
heart_disease_data = no_missing_values[no_missing_values['diagnosis']!=0]
```

```python
convertColumnToInt(no_missing_values)
convertColumnToInt(no_heart_disease_data)
convertColumnToInt(heart_disease_data)
```

```python
sns.kdeplot(
    data=no_missing_values, x="age", y="chol", hue="diagnosis", fill=True, palette = 'crest',
).set_title('Density Plot of Age vs Cholesterol Separated by Diagnosis');
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
plt.title('3D Scatter of Heart Disease Features')
colors = np.where(no_missing_values.diagnosis > 0, "blue", "cornflowerblue")
ax.scatter(no_missing_values.chol, no_missing_values.max_hr,no_missing_values.age,c=colors)
ax.set_zlabel('Age')
ax.set_xlabel('Cholesterol')
ax.set_ylabel('Max HR')
ax.dist = 13
plt.show()
```
